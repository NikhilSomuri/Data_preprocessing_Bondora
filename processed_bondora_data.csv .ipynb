```python
def handle_missing_values(df):
    for column in df.columns:
        if df[column].dtype in ['float64', 'int64']:  # Numeric columns
            median_value = df[column].median()
            df[column].fillna(median_value, inplace=True)
        else:  # Categorical columns
            most_frequent = df[column].mode()[0]
            df[column].fillna(most_frequent, inplace=True)
    return df
```

```python
# Function to identify outliers using IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)  # First quartile (25th percentile)
    Q3 = df[column].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1                   # Interquartile range
    lower_bound = Q1 - 1.5 * IQR    # Lower bound
    upper_bound = Q3 + 1.5 * IQR    # Upper bound

    # Identify outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Apply the function to numeric columns
numeric_columns = data.select_dtypes(include=['number']).columns
for col in numeric_columns:
    outliers = detect_outliers_iqr(data, col)
```

```python
 #Function to cap extreme values
def cap_extreme_values(df, column, lower_percentile=0.01, upper_percentile=0.99):
    lower_bound = df[column].quantile(lower_percentile)  # 1st percentile
    upper_bound = df[column].quantile(upper_percentile)  # 99th percentile
    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)  # Cap values

# Apply the capping function to all numeric columns
numeric_columns = data.select_dtypes(include=['number']).columns
for col in numeric_columns:
    cap_extreme_values(data, col)
```
